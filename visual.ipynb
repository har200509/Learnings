```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

```python
# Bivariate analysis : analysis on two columns at the same time
## 1. Scatterplot (Numerical - Numerical)
sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=tips)
```

```python
# Lineplot (numercial-numercial)
# use when you have a time based quantity (plot on x axis)
new_df=flights.groupby('year')['passengers'].sum().reset_index()
sns.lineplot(x='year',y='passengers',data=new_df)
```

```python
## 2. Bar Plot (Numerical - Categorical)
# x-->category
# y-->numerical
# black rod--> 95% confidence interval for mean
sns.barplot(x='Pclass',y='Age',hue='Sex',data=titanic)
```

```python
## 3. Box Plot (Numerical - Categorical)
# x-->category
# y-->numerical
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

```python
## 4. Displot (Numerical - Categorical)
# filter --> categorical --> numerical
sns.kdeplot(titanic[titanic['Pclass'] == 1]['Age'].dropna(), label='Pclass 1')
sns.kdeplot(titanic[titanic['Pclass'] == 2]['Age'].dropna(), label='Pclass 2')
sns.kdeplot(titanic[titanic['Pclass'] == 3]['Age'].dropna(), label='Pclass 3')
plt.legend()
```

```python
## 5. HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# 6. ClusterMap (Categorical - Categorical)
# similar to heatmap but connects similar maps
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
# Multivariate
## 7. Pairplot (collection of scatterplots) [multiple numerical columns]
sns.pairplot(iris,hue='species')
```
